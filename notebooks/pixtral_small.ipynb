```python
from langchain_mistralai import ChatMistralAI
from IPython.display import display, Image

llm = ChatMistralAI(
    model="pixtral-12b-2409",
    api_key=""# Pixtral-12B model name
)

# List of placeholder images
image_urls = [
    "https://picsum.photos/id/237/800/600",
    "https://picsum.photos/id/1001/800/600",
    "https://picsum.photos/id/1003/800/600",
    "https://picsum.photos/id/1015/800/600",
]

# Loop through images and describe them
for url in image_urls:
    message = [
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Describe this image in detail."},
                {"type": "image_url", "image_url": url},
            ],
        }
    ]
    response = llm.invoke(message)

    # Show image
    display(Image(url=url))

    # Show modelâ€™s description
    print("Description:", response.content, "\n")

```
